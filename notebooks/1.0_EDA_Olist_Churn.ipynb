**Data Joining:** Merge tables to create a comprehensive transactional view per order.

```python
# Example join
df_merged = pd.merge(orders, customers, on='customer_id', how='left')
df_merged['order_purchase_timestamp'] = pd.to_datetime(df_merged['order_purchase_timestamp'])
df_merged = pd.merge(df_merged, order_items, on='order_id', how='left')
```

**Explore Distributions:** Histograms for numerical features (e.g., price, freight value), bar plots for categorical features (e.g., product category, customer state).

```python

import seaborn as sns
import matplotlib.pyplot as plt
sns.histplot(df_merged['price'])
plt.title('Distribution of Product Prices')
plt.show()
```
